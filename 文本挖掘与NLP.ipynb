# 分词

统计语言模型分词法

```
假设一个句子S有多种分词方法，为了简单起见，假定有三种：
A1, A2, A3, …, Ak
B1, B2, B3, …, Bm
C1, C2, C3, …, Cn
我们用k,m,n三个不同的下标表示这个句子在采用不同分词结果时词的数目。最好的一种分词方法应该保证分词后这个句子出现的概率最大
P(A1, A2, A3, …, Ak) > P(B1, B2, B3, …, Bm)
并且
P(A1, A2, A3, …, Ak) > P(C1, C2, C3, …, Cn)


使用动态规划方法，以及维特比算法（HMM）
```

```
分词工具：jieba, snowNLP
```

````python
import jieba
````

````python
data =  open(r'test.txt',encoding = 'utf-8',errors ="ignore").readlines()
````

````python
data[:100]
````

````python
def process(data):
    data_filter = filter(lambda s:len(s)>=5,data)
    m1 = map(lambda s : s.strip("\u3000\u3000|\n"),data_filter)
    m2 = map(lambda s: s.replace("\u3000",""),m1)
    cutwords = map(lambda s:list(jieba.cut(s)),m2)
    return list(cutwords)
````

````python
cutwords = process(data)
````
